Python is a commonly used language for scientific analysis.  Python is quite versatile and code can be written in an editor (like Spyder) and run from the terminal in an interactive mode with iPython.  Also, analyses can be shared with colleagues through iPython Notebook.  

There are a number of important libraries for scientific computing, including NumPy, SciPy, and Matplotlib.  In this tutorial, we'll explore using some of these libraries to read in and perform a simple analysis on a solar dataset.  We will be working with the smoothed sunspot number, obtained from NASA Marshall Space Flight Center and originally from the Royal Observatory in Greenwich.  Through the tutorial, we will read data in from the web, plot the data, and fit the sunspot number from solar cycle 23 to determine our own estimate of when the solar maximum occurred.

```python
# We will use a function to obtain some data from the web to work with.  
# Our function will read in the smoothed sunspot number by year from http://solarscience.msfc.nasa.gov/greenwch/spot_num.txt 

def get_sunspotnumber():
    '''
    Get the sunspot number from the NASA one-month smoothed data.
    Pass back the information from the website.
    '''
    
    import urllib2 # import a library that allows us to read data from the web
    import numpy as np # import the NumPy or Numerical Python librar
    
    sunspot_number = urllib2.urlopen('http://solarscience.msfc.nasa.gov/greenwch/spot_num.txt')
    
    sunspot_dtype = [('year', int), ('month', int), ('SSN', float), ('DEV', float)] # the data type for our numpy array
    
    data = np.loadtxt(sunspot_number, skiprows=1, dtype=sunspot_dtype)   # loads the data on the webpage into a numpy array
            
    return data
```

We have now written a function that allows us to download the smoothed sunspot data from a website.  Next, we'll use the function to pass back the data and take a look at what form it is in.

```python
sunspotnumber_data = get_sunspotnumber()

# Let's explore the data.  First, let's find out what the size of our dataset is and what the first and last entries are.
print 'The size of our dataset is: %d' % (len(sunspotnumber_data)) # prints the length of the dataset

print 'The first entry is: ',
print sunspotnumber_data[0]

print 'The last entry is: ',
print sunspotnumber_data[-1]
```

Let's now make a quick plot of the historical data to see how the sunspot number changes throughout the historical record.  In order to do this, it will be useful to convert the dates to a datetime object.  This will become convenient for plotting and working with the data in general.  We will plot the sunspot number ('SSN') versus the date.

```python
import datetime # import the datetime library
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt # import the plotting library

sunspotnumber_dates = [datetime.datetime(x['year'], x['month'], 1) for x in sunspotnumber_data] # convert to datetime objects
sunspotnumber_dates = np.array(sunspotnumber_dates) # convert our Python list to a NumPy array

print sunspotnumber_dates[0] # print the first date

plt.plot(sunspotnumber_dates, sunspotnumber_data['SSN'], 'k-') # plot the SSN versus date as a black (k) line
plt.xlabel('Date')
plt.ylabel('Sunspot Number')

max_SSN_index = np.argmax(sunspotnumber_data['SSN']) 
plt.plot([sunspotnumber_dates[max_SSN_index]], [sunspotnumber_data['SSN'][max_SSN_index]], 'ro', ms=10)
print 'The Maximum Sunspot Number in the data was on: ',
print sunspotnumber_dates[max_SSN_index], 
print ' The number of sunspots recorded: ',
print sunspotnumber_data['SSN'][max_SSN_index]
```

The plot shows the solar cycles over the past few hundred years!  Our current solar cycle is solar cycle 24.  Let's next take a closer look at just the past and current solar cycles.  Solar cycle 23 started in 1996, so we will filter the data to only include dates that are after Jan 1, 1996.  This is easy to do with NumPy arrays and datetime objects.

```python
ind_after1996 = sunspotnumber_dates > datetime.datetime(1996, 1, 1) # Create an index of which dates meet our condition (True) and which do not (False)

print ind_after1996[0], ind_after1996[-1]  # We should see that the first date (in the 1700s!) is False and the most current is True

plt.plot(sunspotnumber_dates[ind_after1996], sunspotnumber_data['SSN'][ind_after1996], 'k-') # Use the index to only plot more recent data
plt.xlabel('Date')
plt.ylabel('Sunspot Number')
```

Now, let's try fitting a simple Gaussian function to the data from each of these solar cycles.  First, the datetime arrays will not be an easy input into our function.  Instead, we'll compute our x-axis as days in the solar cycle.  We will then fit the Gaussian function to the sunspot number data and plot the results.

```python
# Define a function for a Gaussian fit
def fit_gaussian(x, y):
    ''' Fit a Gaussian to the input x and y data.
    Return an array with the best-fit parameters to the data.
    '''
    
    from scipy.optimize import curve_fit # We will load from the scipy library a useful function for fitting a curve
    
    def gauss(x, *p):
        A, mu, sigma = p  # These are input to the data: the normalization, mid-point, and width of the Gaussian
        return A*np.exp(-(x-mu)**2/(2.*sigma**2))
    
    p0 = [100, 5, 5] # Define some initial conditions for first guesses on A, mu, and sigma
    
    coeff, var_matrix = curve_fit(gauss, x, y, p0=p0) # Fit the data
    
    print 'The best-fit values: ',
    print coeff
    
    return coeff # Pass back the results

# Define a function to convert datetime dates to a decimal of a year for the year in the Solar Cycle
def convert_to_solarcycleyear(dates, startdate):
    '''
    Convert calendar date to year in the solar cycle
    Input: an array of datetime objects and the startdate of the solar cycle
    Output: an array of the deminal year since the beginning of the solar cycle
    '''
    
    ssy = [((x - startdate).total_seconds())/(60.*60.*24.*365.2425) for x in dates]
    return np.array(ssy)
```

Now that we have defined our functions, let's use them to fit the data from solar cycle 23.  We will filter only the dates that correspond to that solar cycle.  It looks like 1996 to 2009 are safe guesses for the approximate start and end of the cycle.  We will next use convert_to_solarcycleyear() to convert the dates.  Then we can use this as our x input and the SSN as the y input into our Gaussian function.

```python
ind_solarcycle23 = (sunspotnumber_dates >= datetime.datetime(1996, 1, 1)) & (sunspotnumber_dates <= datetime.datetime(2009, 1, 1))

solarcycle23_cycleyear = convert_to_solarcycleyear(sunspotnumber_dates[ind_solarcycle23], datetime.datetime(1996, 1, 1)) # convert dates to solar cycle year
solarcycle23_SSN = sunspotnumber_data[ind_solarcycle23]['SSN'] # create a new variable for sunspot number of just this cycle

plt.plot(solarcycle23_cycleyear, solarcycle23_SSN, 'k-') # Plot the solar cycle data
plt.xlabel('Solar Cycle Year')
plt.ylabel('Sunspot Number')
plt.title('Solar Cycle 23')
```

```python
fit_parameters = fit_gaussian(solarcycle23_cycleyear, solarcycle23_SSN)
```

```python
# Make a plot of our best-fit parameters and the data

A = fit_parameters[0]
mu = fit_parameters[1]
sigma = fit_parameters[2]

fit_x = np.arange(0, 13, 0.25) # Create a new array that goes from 0 to 13 in steps of 0.25
fit_y = A*np.exp(-(fit_x-mu)**2/(2.*sigma**2))

plt.plot(solarcycle23_cycleyear, solarcycle23_SSN, 'k-', label='Data') # Plot the solar cycle data
plt.plot(fit_x, fit_y, 'r--', label='Fit')
plt.legend()
plt.xlabel('Solar Cycle Year')
plt.ylabel('Sunspot Number')
plt.title('Solar Cycle 23')

#plt.savefig('SunspotCycle23.png') # if you want to save the figure we just made, uncomment this command
```

We did a pretty good job with our simple fit to the data and found our prediction of a solar max to be 5.15 years from Jan 1, 1996.  An easy way to find the year is to use datetime:

```python
days_from_startofcycle = mu*365.2425 # convert from years to days
solarmax_date = datetime.datetime(1996, 1, 1) + datetime.timedelta(days=int(days_from_startofcycle))
print 'The Estimated Solar Max Date was: %s' % (solarmax_date)
```

Our estimate is pretty good.  The solar max did occur in 2001!  Here is a nice image to check out, which shows snapshots of the Sun throughout the solar cycle from SOHO: http://apod.nasa.gov/apod/ap071203.html
